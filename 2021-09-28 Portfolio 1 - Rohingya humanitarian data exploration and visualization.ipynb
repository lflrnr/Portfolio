### Rohingya Refugees in Cox's Bazar, Bangladesh - Exploring and Visualizing Data
#### https://data.humdata.org/dataset/iscg-4w-influx-cox-s-bazar-bangladesh

The Inter Sector Coordination Group (ISCG) is an umbrella entity of partners and organizations coordinating the provision of services to Rohingya refugees and host communities in the Cox's Bazar area of Bangladesh. Each month, the ISCG collects and publishes data on activities for the affected communities and areas. The data below is for for the month of July 2021 and was downloaded on 2021-09-28 in .xlsx format. The goal is to use Python libraries to explore real world data, generate descriptive statistics and visuals. For such a large and complex dataset, it seems that much effort went into cleaning the data which greatly aided with this portfolio project.

Skills: Python, Pandas, Numpy, Matplotlib, and Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# renamed the xlsx file after download for my own convenience
# original file name is: 210731_ISCG_4W_Consolidated_R68_WEB.xlsx
# https://data.humdata.org/dataset/bf11f6b9-1a2c-424a-ae22-7cd9060d14d7/resource/9496224a-ad09-49f3-b275-0801e060e1dd/download/210731_iscg_4w_consolidated_r68_web.xlsx

fourw = pd.read_excel(
    '2021-09-28 Portfolio 1 - Rohingya humanitarian data.xlsx', 
    sheet_name='4W', index_col = 0,
    header=1)
fourw.head(6)
```

### Calculating the number of Programme Partners, Implementing Partners and the overlap of Programme Partners who at times act as Implementing Partners.

```python
Programme_Partners = fourw['Programme Partner'].unique()
Implementing_Partners = fourw['Implementing Partner'].unique()
# same_orgs = np.sum(Programme_Partners == Implementing_Partners) 
    # above results in deprecation warning probably due to arrays of different length
same_orgs = 0
for partner in Programme_Partners:
    for org in Implementing_Partners:
        if partner == org:
            same_orgs += 1
print("The number of Programme Partners: ", len(Programme_Partners))
print("The number of Implementing Partners: ", len(Implementing_Partners))
print("The number of Programme Partners who at times act as Implementing Partners: ", same_orgs)
```

### Assuming that each row is an activity, some having greater impact or reaching more beneficiaries than others, it will be interesting to see the most frequent actors by activity.
#### For acronym definitions of partners, please refer to the xlsx file's tab titled "Cleaning Table Agencies".

```python
Programme_Partners_count = fourw['Programme Partner'].value_counts().sort_values(ascending=False)
Implementing_Partners_count = fourw['Implementing Partner'].value_counts().sort_values(ascending=False)
print('Top 10 Programme Partners by # of activities:\n',Programme_Partners_count[:10],'\n')
print('Top 10 Implementing Partners by # of activities:\n',Implementing_Partners_count[:10])
```

```python
Programme_Partners_10 = Programme_Partners_count.reset_index()[:10]
Implementing_Partners_10 = Implementing_Partners_count.reset_index()[:10]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), dpi=150)
sns.barplot(ax=ax1, x=Programme_Partners_10['index'], y=Programme_Partners_10['Programme Partner'])
sns.barplot(ax=ax2, x=Implementing_Partners_10['index'], y=Implementing_Partners_10['Implementing Partner'])

ax1.set_xlabel('Programme Partners - Top 10')
ax1.set_ylabel('# of Activities')
ax1.set_xticklabels(ax1.get_xticklabels(),rotation = 45, ha='right')

ax2.set_xlabel('Implementing Partners - Top 10')
ax2.set_ylabel('# of Activities')
ax2.set_xticklabels(ax2.get_xticklabels(),rotation = 45, ha='right')

plt.suptitle('Top 10 Programme and Implementing Partners \nby Gross # of Activities')
plt.tight_layout()
```

```python
df = fourw.copy()
    # Copied the main fourw dataframe for ease of debugging since it ...
    # takes a while to reload the xlsx file for each debug.
```

```python
df = df.rename(columns={'Reached Total Beneciaries (Individuals)': 'total_beneficiaries',
                              'Sector/ Sub-sector of Assistance': 'sector_of_assistance'})
    # columns renamed to facilitate groupby summation below
df['sector_of_assistance'].replace({'CwC':'CwC_community_communications',
                                    'GBV':'GBV_gender_based_violence',
                                    'WASH':'WASH_water_sanit_hygiene',
                                    'NFI/Shelter':'NFI_nonfooditems/shelter'}, inplace=True)
    # rows renamed for readibility when plotting, i.e. defining acronyms
df['total_beneficiaries'] = pd.to_numeric(df['total_beneficiaries'], errors='coerce')
    # blank objects in column casted to numeric to allow for summation without errors
df['total_beneficiaries'].dtype
    # verifed column is now numeric and not object type
```

### How many beneficiaries were reached for all activities?

```python
print('Total beneficiaries reached for all activities:', df['total_beneficiaries'].sum())
print('For context, there were 866,547 Rohingya refugees at the end of 2020 per UNHCR.')
print('Beneficiaries include Rohingya as well as host community members.')
print('It seems possible that a beneficiary may receive benefits from more than one activity.')
```

### What are the top 10 Programme and Implementing Partners by number of beneficiaries reached?

```python
programme_bens = df.groupby(by=['Programme Partner'])['total_beneficiaries'].sum().sort_values(ascending=False)[:10]
implementing_bens = np.round(df.groupby(by=['Implementing Partner'])['total_beneficiaries'].sum().sort_values(ascending=False)[:10], 1)
print('Top 10 Programme Partners by # of beneficiaries reached:\n', programme_bens,'\n')
print('Top 10 Implementing Partners by # of beneficiaries reached:\n',implementing_bens)
```

```python
programme_bens_10 = programme_bens.reset_index()
implementing_bens_10 = implementing_bens.reset_index()
programme_bens_10

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), dpi=150)
sns.barplot(ax=ax1, x=programme_bens_10['Programme Partner'], y=programme_bens_10['total_beneficiaries'])
sns.barplot(ax=ax2, x=implementing_bens_10['Implementing Partner'], y=implementing_bens_10['total_beneficiaries'])

ax1.set_xlabel('Programme Partners - Top 10')
ax1.set_ylabel('# of Beneficiaries Reached')
ax1.set_xticklabels(ax1.get_xticklabels(),rotation = 45, ha='right')
ax1.ticklabel_format(axis='y', style='plain') # avoid auto default to scientific notation on y-axis

ax2.set_xlabel('Implementing Partners - Top 10')
ax2.set_ylabel('# of Beneficiaries Reached')
ax2.set_xticklabels(ax2.get_xticklabels(),rotation = 45, ha='right')

plt.suptitle('Top 10 Programme and Implementing Partners \nby Gross # of Beneficiaries Reached')
plt.tight_layout()
```

### What do the benefeciary demographics look like?

```python
# some minor simplification of column names for ease of dataframe calculations
df = df.rename(columns={"Reached Total Female  (Individuals)": 'client_total_female',
                        "Reached Total Male   (Individuals)": 'client_total_male',
                       "Total Girls (< 18) Beneficiaries": 'client_total_girls',
                       "Total Boys (<18) Beneficiaries": 'client_total_boys',
                       "Total Women (18 -59) Beneficiaries": 'client_total_women',
                       "Total Men (18-59) Beneficiaries": 'client_total_men',
                       "Total elderly Women (>59) Beneficiaries": 'client_total_elderlywomen',
                       "Total elderly Men (>59) Beneficiaries": 'client_total_elderlymen'})

female = round(df['client_total_female'].sum())
male = round(df['client_total_male'].sum())
girls = round(df['client_total_girls'].sum())
boys = round(df['client_total_boys'].sum())
women = round(df['client_total_women'].sum())
men = round(df['client_total_men'].sum())
elder_f = round(df['client_total_elderlywomen'].sum())
elder_m = round(df['client_total_elderlymen'].sum())
```
